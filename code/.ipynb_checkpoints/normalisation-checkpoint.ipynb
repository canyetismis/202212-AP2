```python
#REGEX normalisations

#make letters uppercase
clients['client_name'] = clients['client_name'].str.upper()
clients
```

```python
#replace German characters with ASCII letters
clients['client_name'] = clients['client_name'].str.replace(r'Ä', 'AE', regex=True)
clients['client_name'] = clients['client_name'].str.replace(r'Ö', 'OE', regex=True)
clients['client_name'] = clients['client_name'].str.replace(r'Ü', 'UE', regex=True)
clients['client_name'] = clients['client_name'].str.replace(r'ẞ', 'SS', regex=True)
clients
```

```python
#Handle Separators
clients['client_name'] = clients['client_name'].str.replace(r'\sU.\s|\s&\s', ' UND ', regex=True)
clients
```

```python
#Remove all punctuation and replace with spaces
clients['client_name'] = clients['client_name'].str.replace(r'[^A-Z0-9&]', ' ', regex=True)
clients
```

```python
#Replace KG/Kommanditgesellschaft
clients['client_name'] = clients['client_name'].str.replace(r'\sK\s?G(\s?|$)|KOMMANDITGESELLSCHAFT', ' KG', regex=True)
clients
```

```python
#Replace CO KG
clients['client_name'] = clients['client_name'].str.replace(r'\sC\s?O\s?KG(\s?|$)', ' CO KG', regex=True)
clients
```

```python
#Replace GMBH and handle GMBHUND
regex_str = r'\sG\s?M\s?B\s?H(\s|$)|GESELLSCHAFT\s?M\s?B\s?H(\s|$)|GESELLSCHAFT\s?MIT\s?BESCHRAENKTER\s?HAFTUNG(\s|$)'
clients['client_name'] = clients['client_name'].str.replace(regex_str, ' GMBH', regex=True)
clients['client_name'] = clients['client_name'].str.replace(r'GMBHUND', ' GMBH UND ', regex=True)
clients
```

```python
#Replace AG/AKTIENGESELLSCHAFT
clients['client_name'] = clients['client_name'].str.replace('\sA\s?G(\s|$)|AKTIENGESELLSCHAFT', ' AG', regex=True)
clients
```

```python
# Replace OHG/OFFENE HANDELSGESELLSCHAFT
clients['client_name'] = clients['client_name'].str.replace(r'\sO\s?H\s?G(\s|$)|OFFENE\s?HANDELSGESELLSCHAFT', ' OHG', regex=True)
clients
```
